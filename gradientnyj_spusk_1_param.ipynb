```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2>С одной переменной</h2>

```python
func = lambda x: 3 * x ** 2
diff_func = lambda x: 6 * x
x = np.arange(-10, 11)
func(x)
plt.plot(x, func(x), color='#cc3399', linewidth=2)
```

```python
x_0 = 5
lbd = 0.1
res = []
x_1 = x_0 - lbd * diff_func(x_0)
norm = ((x_1 - x_0) ** 2 + (func(x_1) - func(x_0)) ** 2) ** 0.5
while norm > 0.00000001:
    res.append([x_0, func(x_0)])
    x_0 = x_1
    x_1 = x_0 - lbd * diff_func(x_0)
    norm = ((x_1 - x_0) ** 2 + (func(x_1) - func(x_0)) ** 2) ** 0.5

#
res = np.array(res).reshape(22, 2)
res

x = np.arange(0, 5.1, 0.1)
fig, ax = plt.subplots(figsize=(18, 15))
ax.plot(res[:,0], res[:,1], 'o-')
ax.plot(x, func(x), color='#cc3399', linewidth=3)

```

<h2>С двумя парметрами</h2>

```python
func_2 = lambda x: 3 * x[0] ** 2 + 5 * x[1] ** 2
diff_func_2 = lambda x: np.array([6 * x[0], 10 * x[1]])
x_y_0 = np.array([5, 10])
res_2 = []
x_y_1 = x_y_0 - lbd * func_2(x_y_0)
norm_2 = ((x_y_1[0] - x_y_0[0]) ** 2 + (x_y_1[1] - x_y_0[1]) ** 2 + (func_2(x_y_1) - func_2(x_y_0)) ** 2) ** 0.5
while norm_2 > 0.00000001:
    print(func_2(x_y_0), diff_func_2(x_y_0), sep='----------')
    res_2.append([x_y_0[0], x_y_0[1], func_2(x_y_0)])
    x_y_0 = x_y_1
    x_y_1 = x_y_0 - lbd * diff_func_2(x_y_0)
    norm_2 = ((x_y_1[0] - x_y_0[0]) ** 2 + (x_y_1[1] - x_y_0[1]) ** 2 + (func_2(x_y_1) - func_2(x_y_0)) ** 2) ** 0.5
```

```python
res_2 = np.array(res_2).reshape(25, 3)
res_2
```

```python
X_sfc = np.arange(-10, 10, 0.1)
Y_sfc = np.arange(-10, 10, 0.1)
X_sfc_coords = np.meshgrid(X_sfc, Y_sfc)
X_ft_coords = np.meshgrid(res_2[:,0], res_2[:,1])

Z_sfc = func_2(X_sfc_coords)
Z_ft = func_2(X_ft_coords)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(X_ft_coords[0], X_ft_coords[1], Z_ft)
ax.plot_surface(X_sfc_coords[0], X_sfc_coords[1], Z_sfc, alpha=0.33)
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(0, 600)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
